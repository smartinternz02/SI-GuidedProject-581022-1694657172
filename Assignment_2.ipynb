## Importing Libraries

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

## Loading dataset

```python
data = sns.load_dataset("car_crashes")
data
```

```python
data.info()
```

## First five and last five rows of the dataset

```python
data.head()
```

```python
data.tail()
```

## **Scatter Plot**

```python
sns.scatterplot(y="speeding",x="total",data=data)
```

### **Inference** : As speeding increases total crashes increases from the plot

```python
sns.scatterplot(y="alcohol",x="total",data=data)
```

### **Inference** : As alcohol increases total crashes increases from the plot

## Line Plot


```python
sns.lineplot(x="no_previous",y="not_distracted",data=data)
```

### **Inference** : Overall from the graph as no_previous increases not_distributed increases  

## Distribution plot

```python
sns.distplot(data["total"])
```

### **Inference** : The density of the plot is highest in the range of 10-20

```python
sns.relplot(x="total",y="alcohol",data=data,hue="speeding")
```

### **Inference** : From the plot total increases as alcohol and speeding increases

## Bar Plot

```python
plt.subplots(figsize=(20,10))
sns.barplot(data=data,x="total",y="alcohol",ci=None)
```

### **Inference** : The above plot shows the same relation that alcohol increases total also increases

```python
plt.subplots(figsize=(20,10))
sns.countplot(x="total",data=data)
```

### **Inference** : From the above plot total count of most occurences is 1 time except 6 values which occured twice

## Joint Plot

```python
sns.jointplot(x="total",y="speeding",data=data)
```

### **Inference** : The above plot shows that the frequency of occurence is maximum at the middle values of both parameters

## Box Plot

```python
plt.subplots(figsize=(20,10))
sns.boxplot(x="total",y="alcohol",data=data)
```

### **Inference** : From the above plot, The values are medium values in range of 3 to 6, maximum is 7, minimum is 1

## Correlation and heatmap

```python
corr = data.corr()
corr
```

### **Inference** : The above plot shows correlation between all variables

```python
sns.heatmap(corr,annot=True,cmap="YlGnBu")
```

### **Inference** : The above plot, no_previous and alcohol are more correlated (positive co-relation)
